```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
import warnings
warnings.filterwarnings("ignore")
```

```python
# Load datasets
train = pd.read_csv("train.csv", parse_dates=['date'])
val = pd.read_csv("val.csv", parse_dates=['date'])
test = pd.read_csv("test.csv", parse_dates=['date'])
```

```python
# Ensure sorted
train = train.sort_values("date")
val = val.sort_values("date")
test = test.sort_values("date")

print("Regions:", train['Region'].unique())
```

```python
# Sequence preparation
def create_sequences(data, seq_length=7):
    X, y = [], []
    for i in range(len(data)-seq_length):
        X.append(data[i:i+seq_length])
        y.append(data[i+seq_length])
    return np.array(X), np.array(y)

seq_length = 7
results = []
```

```python
# Train LSTM per region
for Region in train['Region'].unique():
    print(f"\nüîπ Training LSTM for region: {Region}")

    # Use daily averages
    train_series = train[train['Region']==Region].groupby('date')['usage_cpu'].mean().values.reshape(-1,1)
    val_series = val[val['Region']==Region].groupby('date')['usage_cpu'].mean().values.reshape(-1,1)

    # Scale
    scaler = MinMaxScaler()
    train_scaled = scaler.fit_transform(train_series)
    val_scaled = scaler.transform(val_series)

    # Sequences
    X_train, y_train = create_sequences(train_scaled, seq_length)
    X_val, y_val = create_sequences(val_scaled, seq_length)

    # Reshape for LSTM [samples, timesteps, features]
    X_train = X_train.reshape((X_train.shape[0], X_train.shape[1], 1))
    X_val = X_val.reshape((X_val.shape[0], X_val.shape[1], 1))

    # LSTM Model
    model = Sequential()
    model.add(LSTM(50, activation='relu', input_shape=(seq_length,1)))
    model.add(Dense(1))
    model.compile(optimizer='adam', loss='mse')

    model.fit(X_train, y_train, validation_data=(X_val, y_val), epochs=20, batch_size=16, verbose=1)

    # Predictions
    y_pred = model.predict(X_val)
    y_pred_inv = scaler.inverse_transform(y_pred)
    y_val_inv = scaler.inverse_transform(y_val.reshape(-1,1))

    # Metrics
    mae = mean_absolute_error(y_val_inv, y_pred_inv)
    rmse = np.sqrt(mean_squared_error(y_val_inv, y_pred_inv))
    mape = np.mean(np.abs((y_val_inv - y_pred_inv) / y_val_inv)) * 100

    results.append({
        "Model": "LSTM",
        "Region": Region,
        "MAE": round(mae,2),
        "RMSE": round(rmse,2),
        "MAPE": f"{mape:.2f}%"
    })

    # Plot
    plt.figure(figsize=(10,5))
    plt.plot(y_val_inv, label="Actual", marker='o')
    plt.plot(y_pred_inv, label="LSTM Forecast", marker='x')
    plt.title(f"LSTM Forecast vs Actual - {Region} (Validation Set)")
    plt.xlabel("Time Steps")
    plt.ylabel("CPU Usage")
    plt.legend()
    plt.show()
```

```python
# -------------------------------
# Final Test Forecast per Region
# -------------------------------
test_results = []
all_forecasts = []

for Region in train['Region'].unique():
    print(f"\nüîÆ Final Test Forecast for region: {Region}")

    # Combine train + val
    train_val_series = pd.concat([
        train[train['Region']==Region],
        val[val['Region']==Region]
    ]).groupby('date')['usage_cpu'].mean().values.reshape(-1,1)

    test_series = test[test['Region']==Region].groupby('date')['usage_cpu'].mean().values.reshape(-1,1)

    # Scale
    scaler = MinMaxScaler()
    train_val_scaled = scaler.fit_transform(train_val_series)
    test_scaled = scaler.transform(test_series)

    # Sequences
    X_train_val, y_train_val = create_sequences(train_val_scaled, seq_length)
    X_test, y_test = create_sequences(test_scaled, seq_length)

    X_train_val = X_train_val.reshape((X_train_val.shape[0], X_train_val.shape[1], 1))
    X_test = X_test.reshape((X_test.shape[0], X_test.shape[1], 1))

    # LSTM Model (retrain on train+val)
    model = Sequential()
    model.add(LSTM(50, activation='relu', input_shape=(seq_length,1)))
    model.add(Dense(1))
    model.compile(optimizer='adam', loss='mse')

    model.fit(X_train_val, y_train_val, epochs=20, batch_size=16, verbose=0)

    # Predictions on test
    y_pred = model.predict(X_test)
    y_pred_inv = scaler.inverse_transform(y_pred)
    y_test_inv = scaler.inverse_transform(y_test.reshape(-1,1))

    # Metrics
    mae = mean_absolute_error(y_test_inv, y_pred_inv)
    rmse = np.sqrt(mean_squared_error(y_test_inv, y_pred_inv))
    mape = np.mean(np.abs((y_test_inv - y_pred_inv) / y_test_inv)) * 100

    test_results.append({
        "Model": "LSTM",
        "Region": Region,
        "MAE": round(mae,2),
        "RMSE": round(rmse,2),
        "MAPE": f"{mape:.2f}%"
    })

    # Store forecast for CSV
    forecast_dates = test[test['Region']==Region]['date'].unique()[seq_length:]
    all_forecasts.extend(zip([Region]*len(forecast_dates), forecast_dates, y_pred_inv.flatten()))

    # Plot
    plt.figure(figsize=(10,5))
    plt.plot(y_test_inv, label="Actual", marker='o')
    plt.plot(y_pred_inv, label="LSTM Forecast", marker='x')
    plt.title(f"LSTM Forecast vs Actual - {Region} (Test Set)")
    plt.xlabel("Time Steps")
    plt.ylabel("CPU Usage")
    plt.legend()
    plt.show()

# Save test metrics
df_test_results = pd.DataFrame(test_results)
df_test_results.to_csv("metrics_lstm_test.csv", index=False)

# Save forecasts
forecast_df = pd.DataFrame(all_forecasts, columns=["Region","Date","Forecasted_CPU_Usage"])
forecast_df.to_csv("lstm_test_forecast.csv", index=False)

print("\n‚úÖ Final Test Forecast Completed!")
print(df_test_results)
```
