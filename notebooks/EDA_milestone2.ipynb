```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load feature-engineered dataset
df = pd.read_csv("final_feature_engineered_dataset.csv")  # or your processed file

print("Shape:", df.shape)
df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns

plt.figure(figsize=(14, 8))
sns.heatmap(df[numeric_cols].corr(), annot=True, cmap='coolwarm')
plt.title('Feature Correlation Heatmap')
plt.show()
```

```python
plt.figure(figsize=(14, 6))
plt.plot(df['date'], df['usage_cpu'], label='CPU Usage')
plt.plot(df['date'], df['usage_storage'], label='Storage Usage')
plt.title('CPU and Storage Usage Trend Over Time')
plt.xlabel('Date')
plt.ylabel('Usage')
plt.legend()
plt.show()
```

```python
sns.histplot(df['usage_cpu'], kde=True)
plt.title('CPU Usage Distribution')
plt.show()
```

```python
sns.histplot(df['usage_storage'], kde=True)
plt.title('Storage Usage Distribution')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='month', y='usage_cpu', data=df)
plt.title('CPU Usage Seasonality by Month')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='day_of_week', y='usage_cpu', data=df)
plt.title('CPU Usage by Day of Week')
plt.show()
```
